```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp(text)
```

```python
tokens = [token.text.lower() for token in doc if not token.is_stop and not token.is_punct and token.text != '\n' and token.text != '\n\n']
```

```python
tokens
```

```python
from collections import Counter
```

```python
word_freq = Counter(tokens)
word_freq
```

```python
max_freq = max(word_freq.values())
for word in word_freq.keys():
    word_freq[word] /= max_freq

word_freq
```

```python
sent_token = [sent.text for sent in doc.sents]
sent_token
```

```python
sent_score = {}

for sent in sent_token:
    for word in sent.split():
        if word.lower() in word_freq.keys():
            if sent not in sent_score.keys():
                sent_score[sent] = word_freq[word]
            else:
                sent_score[sent] += word_freq[word]

sent_score
```

```python
import heapq

num_sentences = len(sent_token)
top_sent = heapq.nlargest(num_sentences//3, sent_score, key=sent_score.get)
```

```python
summary = "".join(top_sent)
print(summary)
```
